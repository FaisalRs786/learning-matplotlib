# FILL BETWEEN PLOT IN MATPLOTLIB

FILL BETWEEN PLOT USED TO FILL GRAPH BETWEEN TWO POINTS AS PLT.FILL_BETWEEN(X= [2,4] , x1  = 2, y2 = 4)

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)

# plt.fill_between(x,y,color = 'r')
# OR

# plt.fill_between(x = [2,4], y1 = 6 , y2 = 8)
# OR

# or using where parameter as
x_array = np.array(x)
plt.fill_between(x,y, where =(x_array>=2)&(x_array<=4))

plt.show()
```
