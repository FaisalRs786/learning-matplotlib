# BARPLOT
**plt.bar(x,y)**

```python
# simple bar plot according to x and y
import matplotlib.pyplot as plt
x = ["c","c++","php","java","python"]
y = [10,20,30,40,50]

# use different parameter to decor graph as
plt.xlabel("languages")
plt.ylabel("popularity")
plt.title("popularity between programming languages")

# use different parameter as :
plt.bar(x,y, width = 0.5 , color = 'g', align = "center", edgecolor = "r", linewidth = 2, linestyle = ":", alpha = 0.7 ,
        label = "popularity and language",)
plt.legend(loc = 9)
plt.show( )
```

**MULTIPLE GRAPH**

```python
# multiple bar graph as:
x = [1,2,3,4,5]
z = [10,20,30,40,50]
y = [20,40,60,80,100]
plt.bar(x,y, color = 'r')
plt.bar(x,z, color = 'b')
plt.show( )

# but these graph are intersect each other we want to create graph bar one by one in a single graph as
```

**MULTIPLE GRAPH IN SINGLE GRAPH AS:**

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5]
z = [10,20,30,40,50]
y = [20,40,60,80,100]

width = 0.2

x_var = np.arange(len(x))
x_var1 = [ i+width for i in x_var ]

plt.bar(x_var, y, color = 'r' , label= "rank", width = 0.2 )
plt.bar(x_var1 , z, color = 'b', label = "popularity", width = 0.2)
plt.xticks(x_var+width/2, ["c","c++","php","java","python"])
plt.legend(loc = 9)
plt.show( )
```

**HORIZONTAL BAR GRAPH USE plt.barh(x,y)**

```python
import matplotlib.pyplot as plt
x = ["c","c++","php","java","python"]
y = [10,20,30,40,50]

plt.barh(x,y)
plt.show( )
```
