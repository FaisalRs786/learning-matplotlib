# SAVEFIG IN MATPLOTLIB

SAVE GRAPH FIGURE USING PLT.SAVEFIG(NAME) FUNCTION

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2, 4, 5, 9, 7 ]
plt.plot(x,y)

# using savfig( ) function to save figure as :
plt.savefig("linear_graph" , dpi = 2000, facecolor = 'yellow', )

# dpi(dot per inch ) parameter used to set quality of graph saving image
# facecolor parameter used to give margin color to the graph image
# transparent set transparent True to make graph transparent
plt.show()
```
