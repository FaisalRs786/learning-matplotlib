# SUBPLOT IN MATPLOTLIB

**SUBPLOT IN MATPLOTLIB USING PLT.SUBPLOT(NROW, NCOLS, INDEX)**

```python
import matplotlib.pyplot as plt

# we create multiple graphs and merge it in a single graph according to rows and colmns using plt.subplot to give each graph as particular index

x = [ 1, 2, 3, 4]
y = [2,4,6,8]


plt.subplot(2,2,1)
plt.plot(x,y)
# this graph it at 0 index in 2 by 2 subplot

plt.subplot(2,2,2)
plt.stackplot(x,y)

plt.subplot(2,2,3)
plt.step(x,y)

plt.subplot(2,2,4)
plt.boxplot(x)

plt.show()
```
