# WORKING WITH TEXT IN MATPLOTLIB

PLT.TEXT() , PLT.ANOTATE() , PLT.LEGEND()

```python
import matplotlib.pyplot as plt
x = ["c","c++","php","java","python"]
y = [10, 20, 30, 40, 50]

plt.step(x,y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

# using plt.text function to show particular text as particular point in graph
plt.text(2,20, "popular", fontsize = 10, style = "italic", bbox = {'facecolor':'red', 'alpha':0.5, 'pad':5})

# using plt.anotate function to show an arrow at particular point with particular text
plt.annotate("c++", xy = (1,20), xytext = (1,35), arrowprops = dict(facecolor = "black") )

plt.legend(['king plot'], loc = 9, facecolor = 'brown', edgecolor = "b", framealpha = 0.7, shadow =True)

plt.show()
```
