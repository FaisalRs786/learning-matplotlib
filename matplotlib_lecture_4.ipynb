# SCATTER PLOT MATPLOTLIB

```python
# SIMPLE SCATTER GRAPH MATPLOTLIB
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,30,40]
color_var  =  [ 1,2,3,4,5] # it will give different colors to each dots
# color_var = ['red','green','blue','yellow','orange'] it will give different colors to each dots when cmap not use when use cpma use list or int numbers

sizes = [400,300,200,500,300] # it will give different sizes to each dots
# we can give colors as plt.scatter(x,y ,c = colors_var) or we can use cmap property to give colors as

plt.scatter(x,y, c = color_var , cmap = 'BrBG' , s = sizes, marker = 'o', alpha = 1, edgecolors = 'g', linewidth = 2, label = 'king graph')

#plt.colorbar() # it give color bar to the graph
# give label to the color bar as:
var = plt.colorbar()
var.set_label('king colors', fontsize = 15)

plt.title('Scatter Plot', fontsize = 15)
plt.xlabel('X-Axis', fontsize = 15)
plt.ylabel('Y-Axis',fontsize = 15)
plt.grid(True)
plt.legend(loc = 9)
plt.show()
```
