# XTICKS( ), YTICKS( ), XLIM( ), YLIM( ), AXIS( ), IN MATPLOTLIB

**XTICKS( ) AND YTICKS( ) FUNCTION ARE USED TO CHANGE THE X-AXIS AND Y-AXIS LABELS OF GRAPH**  

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,6,8,10,12,14,16,18]

plt.plot(x,y)

# here  we use xticks and yticks function
plt.xticks(x, labels =['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'])
plt.yticks(y, labels =[9,8,7,6,5,4,3,2,1])

plt.show()
```

xlim and ylim are used to set xaxis and yaxis limits as

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = [0.2,0.3,0.6,0.7,1.0,1.4,1.8,2.0,2.2]

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,20)

plt.show()
```

axis function are used to set axis of graph as plt.axis([0,2,0,5])

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,6,8,10,12,14,16,18]

plt.plot(x,y)

plt.axis([0,10,0,20])

plt.show()
```
