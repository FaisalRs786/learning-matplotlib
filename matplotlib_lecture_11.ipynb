# STEP PLOT IN MATPLOTLIB

**STEP PLOT IN MATPLOTLIB USING PLT.STEP(x,y)**

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.step(x,y, marker = "o", color = "red", ms = 5, ls = "--", mfc = "b")
plt.show( )


# marker parameter is used to give markers as marker = "o"
# color parameter used change color of graph
# ls line style
# ms marker size
# mfc marker face color paramter used to change color of marker
```
