BOX PLOT OR WHISKER PLOT IN MATPLOTLIB

**BOX PLOT OR WHISKER PLOT USING MATPLOTLIB AS plt.boxplot(x)**

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x)
plt.show( )
# this is a simple box plot
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x, notch = True , widths= 0.5,patch_artist = True, showmeans = True, whis = 2, sym = "r+", boxprops=dict(color = "r"),
            capprops = dict(color='b'), whiskerprops = dict(color="black"), flierprops = dict(color="r"))
plt.show( )


# notch parameter True to make box shape to notch as noch = True
# vert paramter used to false veritical to make graph horixontal as vert = False
# width parameter used to set width of box in box plot as width = 0.5
# patch_artist parameter used to give color to box in box plot as patch_artist = True
# showmean parameter is used to give point to the mean point or in mean line as show_mean = True
# whis paramter used to expand whisk to the outlier as whis = 2
# sym parameter is used to give color and change shape of outlier as sym = "g+" g is green color and + is shape of outlier
# boxprops parameter use to give different properties to the box as boxprops=dict(color="r")
# capprops use to change color of caps of box in boxplox capprops = dict(color='b')
# whiskerprop used to change color of whisk as whiskerprops = dict(color="r")
# flierprops parameter is used to change color of median line
#
```

# CREATE MULTIPLE BOXES IN SINGLE BOX PLOT AS

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
y = [ 20,40,60,80,100,120,140,200]
plt.boxplot(x)
plt.boxplot(y)
plt.show()

# these boxes merge each other
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
y = [ 20,40,60,80,100,120,140,300]

z = [x,y]
plt.boxplot(z , labels = ["box1", "box2"])
plt.show()
```
