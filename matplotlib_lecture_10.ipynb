# STACK PLOT / AREA PLOT IN MATPLOTLIB

**STACKPLOT IN MATPLOTLIB AS PLT.STACKPLOT( X,Y )**

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
y = [ 20,40,60,80,100,120,140,200]
plt.stackplot(x , y)
plt.show( )
# this a simple area plot / stack plot
```

**MULTIPLE AREA PLOT / STACK PLOT**

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
area1 = [ 20,40,60,50,100,40,80,200 ]
area2 = [ 50,40,80,70,10,100,50,20 ]
area3 = [ 50,70,60,80,40,120,10,40 ]

l = ['area1', 'area2', 'area3'] # list of labels to give this area graph as
plt.stackplot(x, area1, area2, area3, labels = l , colors = ['r','g','y'], baseline = "sym")
plt.legend(loc = 'upper left')
plt.show()

# color parameter to give colors to areas in graph
# baseline parameter used to set from where our first area of graph is started as baseline = "wiggle"

```
