# MATPLOTLIB PIE PLOT

**CREATE MATPLOLIB PYPLOT USING PLT.PIE**

```python
import matplotlib.pyplot as plt
x = [ 10, 20, 30, 40, 50]
y = ['c', 'c++', 'php', 'java', 'python']

# use explode parameter to cut as pieces as create list first which to cut and how much cut as
z = [0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode = z)
plt.show()


# this is simple pie plot  now in other cells using different parameters to decor pie chart
```

```python
import matplotlib.pyplot as plt
x = [ 10, 20, 30, 40, 50]
y = ['c', 'c++', 'php', 'java', 'python']

plt.pie(x, labels=y , autopct = "%0.1f%%", shadow = True, radius = 1.5, labeldistance = 1.2, textprops = {"fontsize":15}, counterclock = False,
        wedgeprops = {'linewidth':3, 'edgecolor':'black'}, rotatelabels = True)


# use autopct parameter to give percetage to each piece of pie chart as autopct = %o.1f%% f for float
# True shadow parameter to give shadow to the pie chart
# radius = 1.5 to increase size of decrease size of pie chart
# labeldistance parameter used to set the label distance from the pie chart as labeldistance = 1.3
# textprops parameter used to give different properties to text of pie chart as textprops={fontsize:15}
# counter clock used to change direction pi chart as counterclock =False
# wedgeprops use to give extra properties to the pie chart as
# rotatelabels use to rotate the labels of pie chart as rotatelabels = True
plt.show()
```

**CREATE DOT PIE CHART AS**

```python
import matplotlib.pyplot as plt
plt.pie([1] , colors = 'r')
plt.show()
```

**CREATE DONUT PIE CHART OR MULTI PIE CHART AS**

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = ['A','B','C','D','E']
x1 = [10,20,30,40,50]
y1 = ['j','i','h','g','f']
plt.pie(x, labels = y)
plt.pie(x1, labels = y1 , radius = 0.6)
plt.show()
```

**CRAETE RING PIE CHART AS**

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = ['A','B','C','D','E']
plt.pie(x, labels = y , radius= 1)
plt.pie([1] ,colors = 'w',  radius = 0.7)

plt.show()
```
